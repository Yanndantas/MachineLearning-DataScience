```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import statsmodels.api as sm

# LEANDRO DIAS VIEIRA: = 0.75 NO CHECKPOINT 1

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MLDS/Aula 2/runner.csv', sep='\t')
# df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MLDS/Aula 2/imc.csv', sep='\t')
# df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MLDS/Aula 2/visits.csv', sep='\t')
df

# Y = B0 + B1 * X1 + E

# o meu X vai ser running_time e o meu Y será o heartbeat
model = LinearRegression(n_jobs=None, fit_intercept=True, copy_X=True)
x = df['running_time'].to_numpy().reshape(-1,1)
y = df['heartbeat'].to_numpy()
model.fit(x, y)

# intercept_ = B0 = alfa = tetha0
# coef_ = lista com todos os betas a partir do 1;
print(model.score(x, y))
print(model.intercept_)
print(model.coef_)

plt.scatter(x, y, color='red')
plt.plot(df[['running_time']].to_numpy(), model.predict(x))
plt.show()

# R2 = 0.4
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
# ORDINARY LEAST SQUARES = MÍNIMOS QUADRADOS ORDINÁRIOS
model.summary()

# Y = 285 - 3,9 * X
# R2 40%
# Tenho significância estatística

# VAMOS IMAGINAR QUE EU TENHA 1000 OBSERVAÇÕES
# DIVIDIR A MINHA AMOSTRA
# 500 OBSERVAÇÕES PARA GERAR O MODELO
# 500 OBSERVAÇÕES PARA TESTAR O MODELO

# r2 = 0.4
# r2 = 0.38


```
