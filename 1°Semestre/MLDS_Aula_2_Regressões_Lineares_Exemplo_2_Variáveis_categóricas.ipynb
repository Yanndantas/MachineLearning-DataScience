```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import statsmodels.api as sm

# SILVIA LIMA: 0.5 NO CHECKPOINT 1
# NAKANISHI: 0.25 NO CHECKPOINT 1

# df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MLDS/Aula 2/runner.csv', sep='\t')
# df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MLDS/Aula 2/imc.csv', sep='\t')
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MLDS/Aula 2/visits.csv', sep='\t')
df['date'] = pd.to_datetime(df['date'])
df['dow'] = df['date'].dt.dayofweek
df['mon'] = df.apply(lambda row: 1 if row['dow']==0 else 0, axis=1) # monday = segunda-feira
df['tue'] = df.apply(lambda row: 1 if row['dow']==1 else 0, axis=1) # tuesday = terça-feira
df['wed'] = df.apply(lambda row: 1 if row['dow']==2 else 0, axis=1) # wednesday = quarta-feira
df['thu'] = df.apply(lambda row: 1 if row['dow']==3 else 0, axis=1) # thursday = quinta-feira
df['fri'] = df.apply(lambda row: 1 if row['dow']==4 else 0, axis=1) # friday = sexta-feira
df['sat'] = df.apply(lambda row: 1 if row['dow']==5 else 0, axis=1) # saturday = sábado
df['sun'] = df.apply(lambda row: 1 if row['dow']==6 else 0, axis=1) # sunday = domingo

model = LinearRegression(n_jobs=None, fit_intercept=True, copy_X=True)
x = df[['mon','tue','wed','thu','fri','sat','sun']].to_numpy()
y = df['visits']
model.fit(x, y)
model.score(x, y)

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
model.summary()


# Y = PLAQUETAS
# X1 = MEDICACAO A
# X2 = MEDICACAO B
# X3 = MEDICACADO C
#
# Y = BETA0 + BETA1 * X1 + BETA2 * X2 + BETA3 * X3
#
# 100 = BETA0
# 20 = BETA1
# 40 = BETA2
# -30 = BETA3

# Y = 100
#  + 20*0
#  + 40*0
#  - 30*1

# INCREMENTO DE PLAQUETAS PARA A MEDICAÇÃO A É DE 20
# INCREMENTO DE PLAQUETAS PARA A MEDICAÇÃO B É DE 40
# INCREMENTO DE PLAQUETAS PARA A MEDICAÇÃO C É DE -30






# Y = 100 + 20*X1 + 40*X2 -30*X3
# Y = 100 + 20*1 + 40*0 - 30*0
# Y = 100 + 20 + 0 + 0
# Y = 120

# Y = 100 + 20*X1 + 40*X2 -30*X3
# Y = 100 + 20*0 + 40*0 - 30*1
# Y = 100 + 0 + 0 - 30*1
# Y = 70


```
